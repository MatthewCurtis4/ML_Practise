# Encoding

```python
#data.select_dtypes('object') returns all columns of type object
#data.select_dtypes('object').columns returns all said column names
{column: len(data[column].unique()) for column in data.select_dtypes('object').columns}
#Creating a dictionary that maps the number of unique values in object columns
```

```python

## Function: Encode the objects into numerical values that only have two options (Sex and Season)
#positive_values: the value option in the column that we define as the positive value when we call it.
def binary_encode(df, columns, positive_values):
    df = df.copy()
    for column, positive_value in zip(columns, positive_values):  #go through both the column value and positive_values
        df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0) #assign the rows column value a 1 if it is the positive_value, 0 otherwise
    return df

## Function: Encode the object values into numerical values that have more than 2 options.
# Basically they will just be represented by the column titles prefix.
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes): #iterate through columns and prefixs this time
        dummies = pd.get_dummies(df[column], prefix=prefix)#get the hot columns for a given column
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = binary_encode(
    data,
    columns = ['Sex', 'Season'],
    positive_values = ['M', 'Summer']
)
data = onehot_encode(
    data,
    columns = ['Team', 'NOC', 'City', 'Sport', 'Event'],
    prefixes= ['T', 'N', 'C', 'S', ' E']
)
```
